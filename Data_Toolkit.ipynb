Q1. What is NumPy, and why is it widely used in Python?
 - NumPy is a Python library used for numerical computations. It provides support for large, multi-dimensional arrays and matrices along with a collection of mathematical functions. It is widely used because it's fast, efficient, and forms the backbone of many data science and machine learning libraries.

Q2. How does broadcasting work in NumPy?
 - Broadcasting allows NumPy to perform arithmetic operations on arrays of different shapes. Instead of manually resizing arrays, NumPy automatically expands them so the operation can be done element-wise.

Q3. What is a Pandas DataFrame?
 - A DataFrame is a two-dimensional, tabular data structure in Pandas with labeled rows and columns. It is similar to a spreadsheet or SQL table and is commonly used for data analysis and manipulation.

Q4. A Explain the use of the groupby() method in Pandas?
 - The groupby() method in Pandas is used to split data into groups based on some criteria, apply a function to each group, and then combine the results. It's helpful for aggregation and summarization.

Q5. Why is Seaborn preferred for statistical visualizations?
 - Seaborn is built on top of Matplotlib and makes beautiful, informative visualizations with less code. It includes built-in support for many types of statistical plots like boxplots, violin plots, and regression plots.

Q6. What are the differences between NumPy arrays and Python lists?
 - NumPy arrays are faster and consume less memory than Python lists. Arrays support vectorized operations and are homogeneous (all elements of the same type), while lists can contain elements of different types.

Q7.What is a heatmap, and when should it be used?
 - A heatmap is a visual representation of data using color to show values. It is often used to display correlation matrices or frequency data in a way that highlights patterns or trends.

Q8. A What does the term “vectorized operation” mean in NumPy?
- Vectorized operations are operations that are applied to entire arrays rather than each element individually. This leads to faster and more concise code.

Q9.A How does Matplotlib differ from Plotly?
 - Matplotlib creates static visualizations and is ideal for reports and publications. Plotly, on the other hand, is designed for interactive and web-based visualizations with features like zooming and tooltips.

Q10.What is the significance of hierarchical indexing in Pandas?
 - Hierarchical indexing, or MultiIndexing, allows you to have multiple levels of indexing on rows or columns. This is useful for working with higher-dimensional data in a 2D DataFrame.

Q11. What is the role of Seaborn’s pairplot() function?
 - The pairplot() function creates scatter plots between all pairs of numeric variables in a dataset. It is useful for visualizing relationships and patterns during exploratory data analysis.

Q12. What is the purpose of the describe() function in Pandas?
 - The describe() function generates summary statistics for numerical columns such as mean, standard deviation, min, max, and quartiles.

Q13. A Why is handling missing data important in Pandas?
 - Missing data can lead to incorrect analysis and results. Pandas provides methods to detect, fill, or remove missing values to ensure the dataset remains accurate and reliable.

Q14.  What are the benefits of using Plotly for data visualization?
 - Plotly allows the creation of interactive and dynamic visualizations. It supports zooming, hovering, and exporting visuals, making it great for dashboards and presentations.

Q15. How does NumPy handle multidimensional arrays?
 - NumPy can handle arrays of any dimension and offers various functions to manipulate them such as reshape(), transpose(), and slicing operations.

Q16. What is the role of Bokeh in data visualization?
 - Bokeh is a Python library used for creating interactive plots and dashboards that can be embedded in web applications. It is especially useful for real-time streaming and large datasets.

Q17. Explain the difference between apply() and map() in Pandas.
 - map() is used for element-wise operations on Series, while apply() can be used on both Series and DataFrames to apply more complex functions.

Q18. What are some advanced features of NumPy?
 - Advanced features of NumPy include broadcasting, universal functions (ufuncs), linear algebra operations, masking, and handling structured arrays.

Q19.  How does Pandas simplify time series analysis?
 - Pandas makes it easy to work with time series through features like datetime indexing, resampling, frequency conversion, and rolling window statistics.

Q20.What is the role of a pivot table in Pandas?
 - A pivot table lets you summarize and reorganize data. You can group by categories and apply aggregate functions to derive insights from the dataset.

Q21. A Why is NumPy’s array slicing faster than Python’s list slicing?
 - NumPy arrays use contiguous memory and are optimized in C, making operations like slicing much faster compared to Python lists.

Q22. What are some common use cases for Seaborn?
 - Common use cases include creating heatmaps, distribution plots, regression plots, and multi-variable visualizations for statistical analysis and data exploration.

```python
#Q4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap
import seaborn as sns
import pandas as pd

df = pd.DataFrame({
    'Math': [80, 90, 85, 95],
    'Science': [78, 85, 82, 88],
    'English': [70, 75, 72, 78]
})

corr = df.corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()


```

```python
#Q8. Create a line plot with multiple lines using Matplotlib
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [10, 20, 30, 40]
y2 = [5, 15, 25, 35]

plt.plot(x, y1, label="Line 1")
plt.plot(x, y2, label="Line 2")
plt.title("Multiple Line Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()

```

```python
#Q13.A Create a 3D scatter plot using Plotly.
import plotly.express as px
import pandas as pd

df = pd.DataFrame({
    'X': [1, 2, 3, 4],
    'Y': [10, 20, 30, 40],
    'Z': [5, 15, 25, 35]
})

fig = px.scatter_3d(df, x='X', y='Y', z='Z', title="3D Scatter Plot")
fig.show()
```
